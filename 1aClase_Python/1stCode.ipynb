## Control de flujo

```python
for i in range(10):
    print(i)
for j in range(1,7,2):
    print('j',j) 
lista = ['a','b','c','d']
for letra in lista:
    print(letra)
for letra in lista[::-1]: # para que vaya marcha atrás
    print(letra)
i = 0
while i < 10: # es como un loop pero tiene una condición donde para
    i = i+2
    print (f"i vale {i} ")
```
